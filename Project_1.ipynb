1.The independent variable is whether or not the words are congruent i.e the word condition (congruent vs. incongruent) and the dependent variable is the amount of time taken to perform the test (name the ink colors).

2.An appropriate null hypothesis is that the condition has no impact on the 
mean time required for the test, i.e. the mean time required by the population under the congruent condition is the same as the mean time required by the population under the incongruent condition.
                                µ∆=µ2−µ1
                                H0:µ∆=0

In this case, the alternate hypothesis would be that the mean time taken under the two conditions is not the same.

We could do a dependent T-test on the difference in times between the two 
data conditions: congruent vs. incongruent. The T-test makes sense because we don't know any population statistics a priori, we only have our samples. Because the hypothesis does not state a direction, we will use a two-tailed test- we simply want to show that the condition has an impact on time required, regardless of whether it's slower or faster.
                                HA:µ∆≠0

    4. Note in the plot below that for every single participant, their time under the congruent condition is less than their time under the incongruent condition

```python
difference = data['Congruent']-data['Incongruent']
```

```python
#Here we compute the standard deviation of the 
#differences in time between the two conditions.
#Specifying ddof=1 causes a denominator of (n-1)
#to be used
S = difference.std(ddof=1)
print S
```

```python
x_congruent = data['Congruent'].mean()
x_incongruent = data['Incongruent'].mean()
n = len(data)
#our point estimate for mu_congruent - mu_incongruent is:
x_congruent-x_incongruent
```

```python
#t = (mu_congruent - mu_incongruent)/(S/sqrt(n))
t = (x_congruent - x_incongruent)/(S/(n**.5))
```

```python
print t
```

    6. I suspect the delay observed in the incongruent condition is due to the cognitive interference caused by the Stroop task. Experiment participants are being asked to deal with two competing stimuli and resolving the conflict between the two requires thought and thus additional time. 

    Another task which might yield similar results would be flash cards with numbers printed on them of unequal size. In the congruent condition, the numerically larger value would also be printed physically larger, i.e. 7 is bigger than 2.

    In the incongruent condition, the numerically smaller value is printed larger than the numerically larger value, i.e. 2 is bigger than 7. If a subject is then asked which number is "bigger" they might respond more slowly in the incongruent condition.
